```python
import pandas as pd
```

```python
df = pd.read_csv("endorsements-2020.csv")
```

```python
df.head()
```

```python
grid= pd.read_json("gridmap-layout-usa-538.json", orient='columns')
```

```python
grid.head()
```

```python
df_merged = df.merge(grid, left_on = "state", right_on = "key")
```

```python
df_sum = df_merged.groupby("state")["points"].sum().reset_index()
df_merged_sum = df_merged.merge(df_sum, left_on = "state", right_on = "state")


```

```python
df_merged_sum.columns = ["date", "position", "city", "state", "endorser", "endorsee", "endorser party", "source", "order", "category", "body", "district", "points", "key", "name", "x", "y", "total_points"]
```

```python
df_merged_sum
```

```python
df_merged_sum.head()
```

```python
candidate_sum = df_merged_sum.groupby(["endorsee", "state"])["points"].sum().reset_index()
```

```python
df_merged_sum_share = df_merged_sum.merge(candidate_sum, left_on = ["endorsee", "state"], right_on = ["endorsee", "state"])
```

```python
df_merged_sum_share.columns = ["date", "position", "city", "state", "endorser", "endorsee", "endorser party", "source", "order", "category", "body", "district", "points", "key", "name", "x", "y", "total_points", "points_endorsee_state"]
```

```python
df_merged_sum_share["share_points"] = df_merged_sum_share["points_endorsee_state"] / df_merged_sum_share["total_points"]
```

```python
grid_data = grid.merge(df_merged_sum_share, left_on = "key", right_on = "state")
```

```python
pd.set_option("display.max_columns", 40)
```

```python
grid_data.drop(["name_y", "key_y", "x_y", "y_y"], 1, inplace = True)
```

```python
grid_data.columns = ["key", "name", "x", "y", "date", "position", "city", "state", "endorser", "endorsee", "endorser party", "source", "order", "category", "body", "district", "points", "total_points", "points_endorsee_state", "share_points"]
```
